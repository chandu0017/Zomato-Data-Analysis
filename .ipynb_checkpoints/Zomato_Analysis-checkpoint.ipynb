```python
df = pd.read_csv('zomato.csv', encoding='latin1')
df.head(2)
```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()

# Rename if necessary
df.rename(columns={'Rate': 'rate', 'Rating': 'rate'}, inplace=True)
```

```python
print(df.columns)
```

```python
df.rename(columns={
    'Aggregate rating': 'rate',
    'Average Cost for two': 'approx_cost_for_two'
}, inplace=True)
```

```python
df['rate'] = df['rate'].replace({'NEW': np.nan, '-': np.nan}).astype(float)
```

```python
# Check null values
print(df.isnull().sum())

# Fill or drop
df['Cuisines'].fillna('Unknown', inplace=True)
df.dropna(subset=['rate', 'approx_cost_for_two'], inplace=True)
```

```python
# Ensure column exists and clean it
if 'rate' in df.columns:
    # Convert to string to handle any non-numeric cases
    df['rate'] = df['rate'].astype(str)

    # Replace unwanted text with NaN
    df['rate'] = df['rate'].replace({'NEW': np.nan, '-': np.nan})

    # Convert back to float safely
    df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
else:
    print("Column 'rate' not found. Make sure you renamed 'Aggregate rating' to 'rate'.")
```

```python
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

üèôÔ∏è Top 10 Cities by Restaurant Count
This code identifies and visualizes the cities with the highest number of restaurants in the dataset.

üîπ What It Does:
Uses .value_counts() on the City column to count restaurant entries per city

Selects the top 10 cities with the most listings

Plots a horizontal bar chart using Matplotlib with a green color palette

üîç Insight:
New Delhi has the highest concentration of restaurants‚Äîover 5000

Followed by Gurgaon and Noida, with significantly fewer entries

Highlights regional density and potential market saturation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read cleaned data
df = pd.read_csv('cleaned_zomato.csv')

# Top 10 cities by restaurant count
top_cities = df['City'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.show()
```

üçΩÔ∏è Top 10 Cuisines
This code identifies and visualizes the most popular cuisines across all restaurants in the dataset.

üîπ What It Does:
Splits the Cuisines column by commas to separate multiple cuisines per restaurant

Uses .explode() to flatten the list into individual entries

Counts the frequency of each cuisine and selects the top 10

Plots a horizontal bar chart using Seaborn with the mako color palette

üîç Insight:
North Indian, Chinese, and Fast Food are the most common cuisines

Helps understand customer preferences and market saturation

```python
# Split cuisines by comma and count top 10
cuisine_series = df['Cuisines'].dropna().str.split(',').explode()
top_cuisines = cuisine_series.value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='mako')
plt.title('Top 10 Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```

ü•ß Online Delivery vs No Online Delivery
This code creates a pie chart to show the proportion of restaurants that offer online delivery versus those that don‚Äôt.

üîπ What It Does:
Uses value_counts() to count how many restaurants have online delivery (Yes) and how many don‚Äôt (No)

Plots those counts as a pie chart with percentage labels (autopct='%1.1f%%')

startangle=140 rotates the chart for better visual balance

Custom colors: blue for "Yes", red for "No"

üîç Insight:
A large majority of restaurants do not offer online delivery

Only about 25‚Äì30% provide online ordering services

This highlights a potential gap in digital service adoption

```python
# Online delivery distribution
online_delivery = df['Has Online delivery'].value_counts()

# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(online_delivery, labels=online_delivery.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999'])
plt.title('Online Delivery vs No Online Delivery')
plt.show()
```

üí∏ Cost for Two vs Rating
This code creates a scatter plot to visualize the relationship between a restaurant‚Äôs cost for two and its rating.

üîπ What It Does:
Plots each restaurant as a dot based on its cost and rating

alpha=0.5 adds transparency to handle overlapping points

Helps identify patterns or clusters in pricing vs customer satisfaction

üì¶ Box Plot of Rating by Price Range
This code creates a box plot to show how restaurant ratings are distributed across different price ranges.

üîπ What It Does:
Groups ratings by Price range

Displays median, quartiles, and outliers for each group

Uses palette='Set2' for a clean color scheme

üîç Insight:
Higher price ranges tend to have higher median ratings

There's more variation in lower price ranges, with some outliers

üìä Average Rating by Price Range
This code calculates the average restaurant rating for each price range and visualizes it using a bar chart.

Insight: Higher price ranges tend to receive better ratings.

Tool Used: groupby() from Pandas and barplot() from Seaborn

Purpose: To explore if expensive restaurants are rated more favorably by customers.

```python
price_rating = df.groupby('Price range')['rate'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='Price range', y='rate', data=price_rating, palette='coolwarm')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()
```
